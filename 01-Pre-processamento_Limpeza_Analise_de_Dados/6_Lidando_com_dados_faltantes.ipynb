# 6 - Lidando com dados faltantes

## Identificando valores faltantes em dados tabulares
Vamos iniciar nosso estudo identificando dados faltantes em tabelas.

Criando um DataFrame 'df' a partir de um arquivo csv usando o método read_csv

```python
import pandas as pd

df = pd.read_csv('missing.csv')
df
```

Usando o método isnull() é possível saber quais as células que tem dados ou não no DataFrame 'df'

Em seguida, usando o método sum é possível saber a quantidade de dados faltantes por coluna

```python
df.isnull().sum()
```

Criando um array numpy com os valores do DataFrame 'df'

```python
# Cria um array numpy com todos os valores do DataFrame

df.values
```

## Eliminando amostras ou *features* com valores ausentes

Removendo linhas do DataFrame que contém ao menos um valor faltante usando o método dropna passando o eixo 0

```python
# Remove colunas que contém valores faltantes

df.dropna(axis=0)
```

Removendo colunas do DataFrame que contém ao menos um valor faltante usando o método dropna passando o eixo 1

```python
# Removendo colunas que contém valores faltantes

df.dropna(axis=1)
```

Removendo as linhas do DataFrame em que todas as colunas contém dados faltantes usando o método dropna e passando como argumento 'how = 'all' '

Como não existe nenhuma linha que atende a essa condição, o DataFrame permanece o mesmo

```python
# "Drop" apenas nas linhas onde todas as colunas são NaN

df.dropna(how='all')
```

Removendo as linhas do DataFrame que não contém ao menos 4 valores reais usando o método dropna e passando como argumento 'thresh = 4'

```python
# Removendo linhas que possuem menos de 3 valores reais

df.dropna(thresh=4)
```

Removendo as linhas do DataFrame que contêm valores faltantes na coluna 'C' usando o método dropna e passando como argumento "subset = ['C'] "

```python
# Remove apenas linhas onde NaN aparece em uma coluna específica

df.dropna(subset=['C'])
```
